```python
from skimage import io, exposure
import skimage.io
import matplotlib.pyplot as plt
import numpy as np
```

```python
def _read_img(image_path):
    img = skimage.io.imread(image_path, plugin='tifffile')
    return img

def image_histogram(img):

    co, ce = exposure.histogram(img)
    
    fig = plt.figure(figsize=(10, 7))
    fig.set_facecolor('white')
    plt.plot(ce[1::], co[1::])
    plt.show()
```

```python
img = _read_img("akhisar_xs_05_04.tif")
```

```python
red_band = img[:,:,0]
NIR_band = img[:,:,-1]

NIR_band = NIR_band.astype(np.float32)
red_band = red_band.astype(np.float32)

# Allow division by zero
np.seterr(divide='ignore', invalid='ignore')

ndvi = (NIR_band - red_band) / (NIR_band + red_band)
```

```python
image_histogram(ndvi)
```
